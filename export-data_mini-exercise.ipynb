```python
'''
Mini Exercise

Diberikan sebuah data imdb_top_1000.csv
Lakukan proses - proses berikut:
Filter data untuk film yang Director-nya adalah Christopher Nolan dan memiliki rating lebih besar dari sama dengan 8.5
Select data dengan kolom - kolom berikut:
"Director"
"Series_Title"
"Released_Year"
"Runtime"
"IMDB_Rating"
Setelah itu, simpan data baru tersebut dengan nama imdb_fix.csv
'''
```

```python
import pandas as pd
```

```python
# read data 
imdb_data = pd.read_csv("D:\\python\\Python for ML n DS\\python_data engineering\\pacmann_datasets\\week 2\\imdb_top_1000.csv") 

imdb_data
```

```python
# filter data based on condition

filtered_imdb_data = imdb_data[(imdb_data["Director"] == "Christopher Nolan") & (imdb_data["IMDB_Rating"]  >= 8.5)]

filtered_imdb_data
```

```python
SELECTED_COLS = ["Director", "Series_Title", "Released_Year", "Runtime", "IMDB_Rating"]

selected_imdb_data = filtered_imdb_data[SELECTED_COLS] 

selected_imdb_data
```

```python
# export data 
selected_imdb_data.to_csv("imdb_fix.csv")
```
