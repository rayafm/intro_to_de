```python
''' 
Mini Exercise 

Diberikan sebuah data:
bank_term_deposit.csv
Lakukan:
Label Encoding pada kolom marital
Simpan hasil encoding ke kolom marital_label
One Hot Encoding pada kolom education
Data Binning berdasarkan kolom age dengan ketentuan:
0 - 5 = Toddler
6 - 11 = Child
12 - 25 = Young Adults
26 - 45 = Middle-aged Adults
46 - 100 = Old-aged Adults
'''
```

```python
# Import library
import pandas as pd
```

```python
# Read data bank_term_deposit.csv
deposit_df = pd.read_csv("D:\\python\\Python for ML n DS\\python_data engineering\\pacmann_datasets\\week 3\\bank_term_deposit.csv")

# View 5 top data
deposit_df.head()
```

```python
# View data type of each column
deposit_df.dtypes
```

```python
### Label Encoding based on `marital` column ###
# Casting data
deposit_df['marital'] = deposit_df['marital'].astype('category')
```

```python
# View data type of each column
deposit_df.dtypes
```

```python
# Generate label from 'marital' column and save it into new column 'marital_label'
deposit_df['marital_label'] = deposit_df['marital'].cat.codes

# View 5 top data
deposit_df.head()
```

```python
### One Hot Encoding based on `education` column ###
deposit_df_encoded = pd.get_dummies(deposit_df, columns=["education"])

# View 5 top data
deposit_df_encoded.head()
```

```python
### Data Binning based on `age` column ###
# Range of value 
custom_bins = [0, 5, 11, 25, 45, 100]

# Label 
custom_labels = ['Toddler', 'Child', 'Young Adults', 'Middle-aged Adults', 'Old-aged Adults']

# Create binning data 
deposit_df['age_category'] = pd.cut(deposit_df['age'], 
                                    bins = custom_bins, 
                                    labels = custom_labels)

# View 5 top data 
deposit_df.head()
```
