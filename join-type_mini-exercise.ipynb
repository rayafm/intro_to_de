```python
# import library 
import pandas as pd
```

```python
# read data reddit_opinion_climate_change.csv
opinion1_data = pd.read_csv("D:\\python\\Python for ML n DS\\python_data engineering\\pacmann_datasets\\week 3\\reddit_opinion_climate_change.csv") 

# view top 5 data 
opinion1_data.head()
```

```python
# read data reddit_opinion_climate_change_author.csv 
opinion2_data = pd.read_csv("D:\\python\\Python for ML n DS\\python_data engineering\\pacmann_datasets\\week 3\\reddit_opinion_climate_change_author.csv") 

# view top 5 data
opinion2_data.head()
```

```python
# set "comment_id" become index 

# in opinion1_data
opinion1_data.set_index('comment_id', inplace = True) 

# in opinion2_data 
opinion2_data.set_index('comment_id', inplace = True)
```
