```python
# import library 
import pandas as pd
```

```python
# read csv data 
netflix_data = pd.read_csv("D:\\python\\Python for ML n DS\\python_data engineering\\pacmann_datasets\\week 3\\Netflix_TV_Shows_and_Movies.csv") 

# view 5 top data
netflix_data
```

```python
# Create object to save the grouping result 
# Grouping based on type and age_certification columns
netflix_type_agecert = netflix_data.groupby(by = ['type', 'age_certification']) 

netflix_type_agecert
```

```python
# View the amount of data for each type and age_certification 
netflix_type_agecert.size()
```
