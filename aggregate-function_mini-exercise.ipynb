```python
# import library 
import pandas as pd
```

```python
# read data bank_term_deposit.csv 
deposit_df = pd.read_csv("D:\\python\\Python for ML n DS\\python_data engineering\\pacmann_datasets\\week 3\\bank_term_deposit.csv") 

# view top 5 data 
deposit_df.head()
```

```python
# view data type of each column 
deposit_df.dtypes
```

```python
# aggregation min and max in age and balance columns 
# aggregation mean in age column 
deposit_df.agg({'age' : ['min', 'max', 'mean'], 'balance' : ['min', 'max']})
```

```python
# view modus from marital and education columns 
deposit_df.agg({'marital' : ['mode'], 'education' : ['mode']})
```
