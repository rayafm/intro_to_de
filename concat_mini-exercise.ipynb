```python
'''
Mini Exercise

Diberikan dua buah data
reddit_opinion_climate_change.csv
reddit_opinion_climate_change_author.csv
Lakukanlah concat pada kedua data tersebut.
Sebelum dilakukan concat, jadikan 'comment_id' sebagai index
'''
```

```python
# import library 
import pandas as pd
```

```python
# read data reddit_opinion_climate_change.csv
opinion = pd.read_csv("D:\\python\\Python for ML n DS\\python_data engineering\\pacmann_datasets\\week 3\\reddit_opinion_climate_change.csv")

# View top 5 data
opinion.head()
```

```python
# read data reddit_opinion_climate_change_author.csv
opinion_author = pd.read_csv("D:\\python\\Python for ML n DS\\python_data engineering\\pacmann_datasets\\week 3\\reddit_opinion_climate_change_author.csv")

# View top 5 data
opinion_author.head()
```

```python
# From reddit_opinion_climate_change.csv data, set "comment_id" become index 
opinion.set_index('comment_id', inplace = True)

# View 5 top data 
opinion.head()
```

```python
# From reddit_opinion_climate_change_author.csv data, set "comment_id" become index 
opinion_author.set_index('comment_id', inplace = True)

# View 5 top data 
opinion_author.head()
```

```python
# Create new dataframe baru to save the concat result
opinion_concat = pd.concat([opinion, opinion_author], 
                            axis = 1, 
                            join = 'inner'
                          )

# View 5 top data the concat result
opinion_concat.head()
```
